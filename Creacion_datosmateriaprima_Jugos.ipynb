```python
import pandas as pd

# Definir las proporciones para cada tipo de jugo, estos son valores completamente aleatorios.
data = {
    'Tipo_Jugo': [
        'Naranja',
        'Pomelo',
        'Naranja, Pomelo y Acerola',
        'Pomelo y Mandarina'
    ],
    'Naranja_Litros': [4, 0, 2, 0],
    'Pomelo_Litros': [0, 4, 2, 3],
    'Acerola_Litros': [0, 0, 1, 0],
    'Mandarina_Litros': [0, 0, 0, 2],
    'Agua_Litros': [1, 1, 0.5, 0.5],
    'Azucar_Kg': [0.3, 0.3, 0.5, 0.4],
    'Vitaminas_g': [10, 10, 10, 10]
}


df_jugos = pd.DataFrame(data)


print(df_jugos)


df_jugos.to_csv("Proporciones_Jugos_Citricos_Actualizado.csv", index=False)


# Mi idea aca es incorporar nuestras predicciones de las ventas con la materia prima que se necesita para cada porducto asi podemos tener un estimado de la materia prima
# Necesaria para cada uno de los productos (Nos vamos a enfocar principalmente en los jugos y en SUPERMIX).
# al final en la pagina web vamos a poder añadir un sector donde veremos la cantidad de materia prima que necesitaremos para producir cada producto durante todo el año.
```

```python
# esto viene a ser un codigo ejemplo que nos sirve como guia para cuando tengamos incorporado con la predicciones de las ventas de cada uno de los jugos
# asi como tenemos las predicciones de cada uno de los SUPERMIX:
# Supongamos que tienes un DataFrame de ventas mensuales
df_ventas = pd.DataFrame({
    'Mes': ['Enero', 'Febrero', 'Marzo'],
    'Naranja_Vendido': [100, 150, 200],
    'Pomelo_Vendido': [200, 250, 300],
    'Naranja_Pomelo_Acerola_Vendido': [50, 70, 90],
    'Pomelo_Mandarina_Vendido': [80, 60, 100]
})

# Calcular la materia prima necesaria
def calcular_materia_prima(df_ventas, df_jugos):
    # Crear una lista para almacenar los resultados
    resultados = []

    for index, row in df_ventas.iterrows():
        # Calcular litros necesarios para cada tipo de jugo
        litros_naranja = (row['Naranja_Vendido'] * df_jugos.loc[df_jugos['Tipo_Jugo'] == 'Naranja', 'Naranja_Litros'].values[0])
        litros_pomelo = (row['Pomelo_Vendido'] * df_jugos.loc[df_jugos['Tipo_Jugo'] == 'Pomelo', 'Pomelo_Litros'].values[0])
        litros_acerola = (row['Naranja_Pomelo_Acerola_Vendido'] * df_jugos.loc[df_jugos['Tipo_Jugo'] == 'Naranja, Pomelo y Acerola', 'Acerola_Litros'].values[0])
        litros_mandarina = (row['Pomelo_Mandarina_Vendido'] * df_jugos.loc[df_jugos['Tipo_Jugo'] == 'Pomelo y Mandarina', 'Mandarina_Litros'].values[0])

        # Calcular agua, azúcar y vitaminas
        agua_necesaria = (row['Naranja_Vendido'] * df_jugos.loc[df_jugos['Tipo_Jugo'] == 'Naranja', 'Agua_Litros'].values[0]) + \
                         (row['Pomelo_Vendido'] * df_jugos.loc[df_jugos['Tipo_Jugo'] == 'Pomelo', 'Agua_Litros'].values[0]) + \
                         (row['Naranja_Pomelo_Acerola_Vendido'] * df_jugos.loc[df_jugos['Tipo_Jugo'] == 'Naranja, Pomelo y Acerola', 'Agua_Litros'].values[0]) + \
                         (row['Pomelo_Mandarina_Vendido'] * df_jugos.loc[df_jugos['Tipo_Jugo'] == 'Pomelo y Mandarina', 'Agua_Litros'].values[0])

        azucar_necesaria = (row['Naranja_Vendido'] * df_jugos.loc[df_jugos['Tipo_Jugo'] == 'Naranja', 'Azucar_Kg'].values[0]) + \
                           (row['Pomelo_Vendido'] * df_jugos.loc[df_jugos['Tipo_Jugo'] == 'Pomelo', 'Azucar_Kg'].values[0]) + \
                           (row['Naranja_Pomelo_Acerola_Vendido'] * df_jugos.loc[df_jugos['Tipo_Jugo'] == 'Naranja, Pomelo y Acerola', 'Azucar_Kg'].values[0]) + \
                           (row['Pomelo_Mandarina_Vendido'] * df_jugos.loc[df_jugos['Tipo_Jugo'] == 'Pomelo y Mandarina', 'Azucar_Kg'].values[0])

        vitaminas_necesarias = (row['Naranja_Vendido'] * df_jugos.loc[df_jugos['Tipo_Jugo'] == 'Naranja', 'Vitaminas_g'].values[0]) + \
                               (row['Pomelo_Vendido'] * df_jugos.loc[df_jugos['Tipo_Jugo'] == 'Pomelo', 'Vitaminas_g'].values[0]) + \
                               (row['Naranja_Pomelo_Acerola_Vendido'] * df_jugos.loc[df_jugos['Tipo_Jugo'] == 'Naranja, Pomelo y Acerola', 'Vitaminas_g'].values[0]) + \
                               (row['Pomelo_Mandarina_Vendido'] * df_jugos.loc[df_jugos['Tipo_Jugo'] == 'Pomelo y Mandarina', 'Vitaminas_g'].values[0])

        # Agregar resultados a la lista
        resultados.append({
            'Mes': row['Mes'],
            'Litros_Naranja': litros_naranja,
            'Litros_Pomelo': litros_pomelo,
            'Litros_Acerola': litros_acerola,
            'Litros_Mandarina': litros_mandarina,
            'Agua_Litros': agua_necesaria,
            'Azucar_Kg': azucar_necesaria,
            'Vitaminas_g': vitaminas_necesarias
        })

    # Crear un DataFrame a partir de los resultados
    df_materia_prima = pd.DataFrame(resultados)
    return df_materia_prima

# Llamar a la función
materia_prima_necesaria = calcular_materia_prima(df_ventas, df_jugos)

# Mostrar resultados
materia_prima_necesaria


#Al final lo que vamos a tener vendria a ser una incoporacion con el forecast de cada uno de los productos y la materia prima necesaria para cada producto.
```
