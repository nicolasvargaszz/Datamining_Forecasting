```python
# Import necessary libraries
import pandas as pd

# Placeholder: Load or define df again (user needs to re-upload if necessary)
# df = pd.read_csv("your_data.csv")  # Uncomment if needed

# Ensure 'PRODUCT' column exists in df
if "PRODUCT" in df.columns and "VENTAS_REALES" in df.columns:
    # Create an empty list to store filtered data
    filtered_data_list = []

    # Iterate over unique products
    for product in df["PRODUCT"].unique():
        # Filter data for the current product
        df_product = df[df["PRODUCT"] == product].copy()

        # Calculate percentiles 5 and 85
        p5 = df_product["VENTAS_REALES"].quantile(0.05)
        p90 = df_product["VENTAS_REALES"].quantile(0.85)

        # Apply filtering
        df_filtered = df_product[(df_product["VENTAS_REALES"] >= p5) & (df_product["VENTAS_REALES"] <= p90)]

        # Store filtered data
        filtered_data_list.append(df_filtered)

    # Concatenate all filtered data
    df_final_filtered = pd.concat(filtered_data_list, ignore_index=True)

    # Save to CSV
    df_final_filtered.to_csv("merged_data.csv", index=False)

```
